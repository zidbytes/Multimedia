# Pendahuluan

## Apa itu Data Multimedia?

Data multimedia adalah informasi yang dikodekan dalam berbagai format untuk merepresentasikan dunia nyata:

- **Audio (1D)**: Sinyal satu dimensi yang berubah terhadap waktu
  - Contoh: musik, suara, speech
  - Representasi: amplitudo vs waktu
  
- **Gambar (2D)**: Matriks nilai intensitas dalam ruang dua dimensi
  - Contoh: foto, ilustrasi, grafik
  - Representasi: intensitas pixel pada koordinat (x,y)
  
- **Video (2D + Waktu)**: Rangkaian frame (gambar) yang ditampilkan berurutan
  - Contoh: film, rekaman, animasi
  - Representasi: frame berubah terhadap waktu dengan frame rate tertentu

### A2. TODO: Muat & Metadata

```python
file_path = 'data/audio_musik_polovich_up_and_down.mp3'
data, sample_rate = sf.read(file_path)
info = sf.info(file_path)

# Metadata
channels = info.channels
frames = info.frames
duration_seconds = frames / float(sample_rate)
channel_label = "Mono" if channels == 1 else "Stereo"
total_samples = frames * channels

print("=== Metadata Dasar Audio ===")
print(f"File path            : {file_path}")
print(f"Sample rate (Hz)     : {sample_rate} Hz")
print(f"Durasi               : {duration_seconds:.2f} detik")
print(f"Jumlah kanal         : {channels} ({channel_label})")
print(f"Jumlah total sampel  : {total_samples}")
```

### A4. TODO: Spectrogram log-dB

**Instruksi:**
Hitung STFT dan tampilkan spectrogram dalam skala log-dB:
- Gunakan parameter standar (n_fft=1024, hop_length=256)
- Tampilkan dengan colorbar
- Label sumbu: waktu (detik) dan frekuensi (Hz)

**Analisis yang diperlukan:**
Jelaskan perbedaan informasi yang didapat dari spectrogram dibanding waveform.
Berdasarkan waveform, informasi yang terlihat hanya sebatas perubahan amplitudo terhadap waktu, seperti pada build-up terlihat peningkatan amplitudo yang berlanjut hingga drop. Namun, detail mengenai instrumen atau vocal sulit diidentifikasi. Sebaliknya, spectrogram memberikan gambaran distribusi frekuensi sepanjang waktu sehingga lebih jelas menunjukkan karakter musik. Misalnya, echo pada build-up tampak semakin jelas menjelang drop, dan vokal terlihat pada rentang frekuensi sekitar 2500–5000 Hz. Dengan demikian, spectrogram menyajikan informasi yang lebih kaya dibandingkan waveform.

---

```python
start_sec = 0
end_sec = 80
start_sample = int(start_sec * sample_rate)
end_sample = int(end_sec * sample_rate)

# Cek channel, ambil channel 1 jika stereo
if data.ndim == 1:
    audio_cut = data[start_sample:end_sample]
else:
    audio_cut = data[start_sample:end_sample, 0]

# Hitung STFT
n_fft = 1024
hop_length = 256
D = librosa.stft(audio_cut, n_fft=n_fft, hop_length=hop_length)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

plt.figure(figsize=(10, 4))
librosa.display.specshow(
    S_db,
    sr=sample_rate,
    hop_length=hop_length,
    x_axis='time',
    y_axis='hz',
    cmap='magma'
)
plt.title(f'Spectrogram (Log-dB) - Channel 1 ({start_sec}-{end_sec} detik)')
plt.xlabel('Waktu (detik)')
plt.ylabel('Frekuensi (Hz)')
plt.colorbar(label='Intensitas (dB)')
plt.tight_layout()
plt.show()
```

### A5. TODO: MFCC

**Instruksi:**
Hitung dan tampilkan minimal 13 koefisien MFCC sebagai heatmap:
- Sumbu X: waktu (frame)
- Sumbu Y: koefisien MFCC (1-13)
- Gunakan colorbar dan judul yang jelas

**Analisis yang diperlukan:**
Interpretasi sederhana: apakah pola MFCC stabil atau berubah-ubah? Apa potensi maknanya?

Dari pola MFCC yang diambil dari detik ke 0-80 polanya cenderung stabil hanya ada beberapa perubahan warna yang kemungkinan efek dari vocal dan instrumen tambahan. Pada area drop pada detik 21-50 warna hampir sangat stabil karena beat dan instrumen yang dihasilkan juga cukup stabil.

---



```python
start_sec = 0
end_sec = 80
start_sample = int(start_sec * sample_rate)
end_sample = int(end_sec * sample_rate)

# Ambil channel 1 jika stereo
if data.ndim == 1:
    audio_cut = data[start_sample:end_sample]
else:
    audio_cut = data[start_sample:end_sample, 0]

# Hitung MFCC
n_mfcc = 13
mfccs = librosa.feature.mfcc(
    y=audio_cut.astype(float),
    sr=sample_rate,
    n_mfcc=n_mfcc,
    n_fft=1024,
    hop_length=256
)

plt.figure(figsize=(12, 5))
librosa.display.specshow(
    mfccs,
    x_axis='time',
    sr=sample_rate,
    hop_length=256,
    cmap='viridis'
)
plt.title(f'Heatmap 13 MFCC (Channel 1, {start_sec}-{end_sec} detik)')
plt.ylabel('Koefisien MFCC (1-13)')
plt.xlabel('Waktu (frame)')
plt.colorbar(label='Nilai MFCC')
plt.tight_layout()
plt.show()
```

### C2. TODO: Baca & Metadata

**Instruksi:**
Baca video dengan OpenCV dan tampilkan metadata:
- Resolusi (Width × Height)
- Frame rate (fps)
- Jumlah total frame
- Durasi (detik)
- Klasifikasi resolusi (HD, Full HD, 4K, dll.)

**Analisis:**
Parameter-parameter di atas sangat penting, dijelaska sebagai berikut.
1. Resolusi : Semakin tinggi resolusi dari sebuah video maka informasi yang bisa diambil juga semakin banyak karena akan semakin tajam dari detail dan kualitas gambarnya. Resolusi yang tinggi juga dapat memudahkan jika video ingi digunakan untuk deteksi objek, face recognition, dan lainnya.
2. Frame Rate : fps menentukan seberapa halus gerakan dari sebuah video, semakin tinggi fps, maka semakin mulus videonya. Fps penting untuk keperluan editing video dan pengalaman penonton.
3. Klasifikasi Resolusi : Penting untuk pengalaman menonton dan kemudahan ketika melakukan deteksi.
Jelaskan pentingnya parameter-parameter tersebut untuk analisis video atau aplikasi tertentu.


---



```python
# Path video
video_path = "data/video_edit_jamal_browner_deadlift.mp4"

# Buka video dengan OpenCV
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Gagal membuka video. Periksa path atau file.")
else:
    # Ambil metadata
    width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps    = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps if fps > 0 else 0

    # Klasifikasi resolusi
    if width >= 3840 and height >= 2160:
        resolusi_class = "4K"
    elif width >= 2560 and height >= 1440:
        resolusi_class = "2K / QHD"
    elif width >= 1920 and height >= 1080:
        resolusi_class = "Full HD"
    elif width >= 1280 and height >= 720:
        resolusi_class = "HD"
    else:
        resolusi_class = "SD"

    # Cetak metadata
    print("Metadata Video:")
    print(f"Resolusi       : {width} x {height}")
    print(f"Frame Rate     : {fps:.2f} fps")
    print(f"Total Frame    : {total_frames}")
    print(f"Durasi         : {duration:.2f} detik")
    print(f"Klasifikasi    : {resolusi_class}")
```

### C3. TODO: Tampilkan 3 Frame (Awal–Tengah–Akhir)

**Instruksi:**
Ambil dan tampilkan 3 frame representatif:
- Frame pertama (index 0)
- Frame tengah (index ~total_frame/2)
- Frame terakhir (index total_frame-1)
- **Konversi BGR→RGB** sebelum ditampilkan
- Subplot dengan judul frame dan timestamp

**Analisis:**
1. Frame Pertama: Frame pertama hanya terlihat layar hitam yang biasanya merupakan intro dari video yang tidak memiliki informasi apapun untuk dianalisis lebih lanjut.
2. Frame Tengah: Terlihat gambar seseorang yang sedang melakukan excercise (Sumo Deadlift). Gambar terlihat stabil jika dibandingkan dengan frame terakhir yang berarti gambar diambil dalam kondisi diam. Informasi yang bisa diambil dan dianalisis lebih lanjut bisa analisis aktivitas apa yang sedang dilakukan dan melakukan analisis postur dari gerakan yang dilakukan apakah sudah tepat.
3. Frame Terakhir: Gambar terlihat blur karena gambar diambil dengan gerakan yang cepat.


---

```python
# Ambil 3 frame representatif
frame_indices = [0, total_frames // 2, total_frames - 1]
frames = []

for idx in frame_indices:
    # Set posisi frame
    cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
    ret, frame = cap.read()

    if ret:
        # Konversi BGR ke RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frames.append(frame_rgb)
    else:
        print(f"Gagal membaca frame {idx}")

# Tutup video capture
cap.release()

# Tampilkan 3 frame dalam subplot
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, (frame, idx) in enumerate(zip(frames, frame_indices)):
    timestamp = idx / fps

    axes[i].imshow(frame)
    axes[i].set_title(f'Frame {idx}\nTimestamp: {timestamp:.2f}s')
    axes[i].axis('off')

# Label untuk setiap posisi
labels = ['Frame Awal', 'Frame Tengah', 'Frame Akhir']
for i, label in enumerate(labels):
    axes[i].text(0.5, -0.1, label, transform=axes[i].transAxes,
                ha='center', va='top', fontweight='bold')

plt.tight_layout()
plt.show()
```
